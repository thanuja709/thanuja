```python
# Initialize the FacetGrid object
g = sns.FacetGrid(df, row="Sex", hue="Sex", aspect=5, height=3)

# # Draw the densities in a few steps
g.map(sns.kdeplot, "Age", shade=True, alpha=1, lw=3.5, bw=.2)
g.map(sns.kdeplot, "Age", color="yellow", lw=2, bw=.2)
g.map(plt.axhline, y=0, lw=2)

# # Define and use a simple function to label the plot in axes coordinates
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, color="yellow", ha="left", va="center", transform=ax.transAxes)
g.map(label, "age")

# Set the subplots to overlap
g.fig.subplots_adjust(hspace=-.25)

# # Remove axes details that don't play well with overlap
g.set_titles("")
g.set(yticks=[])
g.despine(bottom=True, left=True)
```
