```python
x = df.drop(['S. No', 'Category', 'State/UT','Percentage_share', 
                     'Projected_population', 'Rate_cyber_crime_2018',
                     'Total'], axis = 1)
y = df['Total']
```

```python
x.shape
```

```python
y.shape
```

# Modelling the data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

# Decision Tree Model.

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor(max_depth=6)
regressor.fit(X_train, y_train)
```

```python
print("Training Accuracy :", regressor.score(X_train, y_train))
print("Testing Accuracy :", regressor.score(X_test, y_test))
```

# Random forest model.

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
rf_regressor.fit(X_train, y_train)
```

```python
print("Training Accuracy :", rf_regressor.score(X_train, y_train))
print("Testing Accuracy :", rf_regressor.score(X_test, y_test))
```
