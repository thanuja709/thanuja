```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
d = {'sqft': [1000,2000,3000,4000,5000], 'price':[100000,200000,300000,400000,500000]}
df = pd.DataFrame(d)
df
```

```python
df
```

```python
x=df[["sqft"]]
y=df[["price"]]
```

```python
plt.scatter(x,y,marker="*",color="red")
```

```python
reg=LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[7000]])
```

```python
plt.xlabel("sqft",fontsize=20)
plt.ylabel("price in l",fontsize=20)
plt.scatter(df.sqft,df.price,color="green",marker="*")
plt.plot(df.sqft,reg.predict(df[["sqft"]]),color="red")
```

```python
df_pred=pd.read_excel("Linear_pred.xlsx")
```

```python
pred=reg.predict(df_pred)
```

```python
df_pred["sqft"]=pred
```

```python
df_pred.to_csv("predict.csv",index=False)
```
